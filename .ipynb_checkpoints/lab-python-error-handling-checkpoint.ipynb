````python
def initialize_inventory(products):
    inventory = {}

    for product in products:
        valid_input = False

        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))

                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative.")

            except ValueError:
                print("Invalid input.")

    return inventory
````

````python
def get_customer_orders(inventory):
    customer_orders = set()

    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative.")
            break
        except ValueError as error:
            print(f"Error: {error}")

    for _ in range(num_orders):
        valid_product = False

        while not valid_product:
            try:
                product = input(
                    "Enter the name of a product that a customer wants to order: "
                ).lower().strip()

                if product not in inventory:
                    raise ValueError("Product does not exist in inventory.")

                if inventory[product] <= 0:
                    raise ValueError("Product is out of stock.")

                customer_orders.add(product)
                valid_product = True

            except ValueError as error:
                print(f"Error: {error}")

    return customer_orders
````

````python
def calculate_total_price(customer_orders):
    total_price = 0

    for product in customer_orders:
        valid_price = False

        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))

                if price < 0:
                    raise ValueError("Price cannot be negative.")

                total_price += price
                valid_price = True

            except ValueError as error:
                print(f"Error: {error}")

    return total_price
````

````python
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        inventory[product] -= 1

    updated_inventory = {
        product: value
        for product, value in inventory.items()
        if value > 0
    }

    return updated_inventory
````

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = initialize_inventory(products)

customer_orders = get_customer_orders(inventory)

updated_inventory = update_inventory(customer_orders, inventory)

total_price = calculate_total_price(customer_orders)

print("\nUpdated Inventory:")
print(updated_inventory)

print("\nTotal Price:", total_price)
````
